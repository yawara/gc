```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## $y = \sin ( \frac{1}{x} )$

```python
plt.ylim([-1.1, 1.1])

X = np.arange(-1, 1, 0.003)
Y = np.sin(1 / X)
plt.plot(X, Y)

X = np.asarray([2 / (i * np.pi) for i in range(1, 10)] + [-2 / (i * np.pi) for i in range(1, 10)])
Y = np.sin(1 / X)
plt.plot(X, Y, 'ro')

plt.show()
```

## $y = x \sin ( \frac{1}{x})$

```python
X = np.arange(-1, 1, 0.003)
Y = X * np.sin(1 / X)
plt.plot(X, Y)
plt.plot(X, X, '--k')
plt.plot(X, -X, '--k')

X = np.asarray([2 / (i * np.pi) for i in range(1, 10)] + [-2 / (i * np.pi) for i in range(1, 10)])
Y = X * np.sin(1 / X)
plt.plot(X, Y, 'ro')

plt.show()
```

```python
from math import modf

def cantor(x, N=100):
    if x > 1 or x < 0:
        raise Exception
    if x == 0:
        return 0
    if x == 1:
        return 1

    rtv = 0

    tmp = x
    for i in range(1, N):
        tmp, finteger = modf(tmp * 3)
        r = int(finteger) % 3
        if r == 2:
            rtv += 2**(-i)
        elif r == 1:
            rtv += 2**(-i)
            break

    return rtv


def np_cantor(X):
    Y = np.zeros(len(X))
    for i, x in enumerate(X):
        Y[i] = cantor(x)

    return Y


X = np.arange(0, 1, 0.001)
Y = np_cantor(X)
plt.plot(X, Y)

plt.show()
```
