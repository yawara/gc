## 5. 次の$U(x,y)$で与えられるスカラー場を等高線表示し，さらに勾配$\mathrm{grad} U$を重ねて図示せよ．

この解答は[フリーソフトウェア](https://ja.wikipedia.org/wiki/%E3%83%95%E3%83%AA%E3%83%BC%E3%82%BD%E3%83%95%E3%83%88%E3%82%A6%E3%82%A7%E3%82%A2)である[Python](https://www.python.org/)（ぱいそん、と読む）で作成しています。

とくに[SciPy](http://scipy.org/)（さいぱい、と読む）という科学技術計算用のPythonパッケージを利用しています。

***以下、ベクトル場の図示については見やすいように適宜長さを調整しています。***

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### (a). $U(x,y)=x$


```python
X, Y = np.meshgrid(np.linspace(-3, 3, 11), np.linspace(-3, 3, 11))
plt.contour(X, Y, X)
plt.quiver(X, Y, 1, 0)
plt.show()
```

### (b). $U(x,y)=y^{2}$

```python
X, Y = np.meshgrid( np.linspace(-3, 3, 11), np.linspace(-3, 3, 11) )
plt.contour(X, Y, Y**2)
plt.quiver(X, Y, 0, 2*Y)
plt.show()
```

### (c). $U(x,y)=x^{2}+y^{2}$


```python
X, Y = np.meshgrid( np.linspace(-6, 6, 11), np.linspace(-6, 6, 11) )
plt.contour(X, Y, X**2+Y**2)
plt.quiver(X, Y, 2*X, 2*Y)
plt.show()
```

### (d). $U(x,y)=x^{2}-y^{2}$


```python
X, Y = np.meshgrid( np.linspace(-6, 6, 11), np.linspace(-6, 6, 11) )
plt.contour(X, Y, X**2-Y**2)
plt.quiver(X, Y, 2*X, -2*Y)
plt.show()
```

### (e). $U(x,y)=y-x^{2}$

```python
X, Y = np.meshgrid( np.linspace(-2, 2, 11), np.linspace(-2, 2, 11) )
plt.contour(X, Y, Y-X**2)
plt.quiver(X, Y, -2*X, 1)
plt.show()
```
