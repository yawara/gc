```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

## 1. 次のスカラー場の等高線を図示せよ

### a. $x+2y$

```python
def f(x,y):
    return x+2*y

X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
plt.show()
```

### b.  $x^{2}+2y^{2}$

```python
def f(x,y):
    return x**2+2*y**2

X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
plt.show()
```

## 2. 直交座標で表された次のベクトル場を図示せよ．

***以下、ベクトル場の図示については見やすいように長さを適宜短めにしています。***

### a. $f(x,y)=e_{x}$

```python
n = 11
X, Y = np.meshgrid(np.linspace(-3, 3, n), np.linspace(-3, 3, n))
U, V = np.array([[ 1 for i in range(n)] for j in range(11)]), np.array([[ 0 for i in range(n)] for j in range(11)])
plt.quiver(X,Y,U,V)
plt.show()
```

### b. $f(x,y)=-ye_x+x_y$

```python
X, Y = np.meshgrid(np.linspace(-5, 5, 11), np.linspace(-5, 5, 11))
U, V = -Y, X
plt.quiver(X, Y, U, V)
plt.show()
```

## 3. 極座標で表示された次のベクトル場を図示せよ
### a. $f(r,\theta)=e_r = ( \cos \theta, \sin \theta )$

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 11), np.linspace(-3, 3, 11))
X[5][5], Y[5][5] = 1, 1
U, V = X/np.sqrt(X**2+Y**2), Y/np.sqrt(X**2+Y**2)
U[5][5], V[5][5] = 0, 0
plt.quiver(X, Y, U, V)
plt.show()
```

## b. $f(r, \theta)=e_\theta=(-\sin \theta,\cos \theta)$

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 11), np.linspace(-3, 3, 11))
X[5][5], Y[5][5] = 1, 1
U, V = -Y/np.sqrt(X**2+Y**2), X/np.sqrt(X**2+Y**2)
U[5][5], V[5][5] = 0, 0
plt.quiver(X, Y, U, V)
plt.show()
```

## 4. $f = x^2+y^2$に対して(16)を確認せよ

### $\frac{\partial f}{\partial r} = \cos \theta \frac{\partial f}{\partial x}+\sin \theta \frac{\partial f}{\partial y}$について

$f = r^2$より、

（左辺）$= 2r$

$x = r \cos \theta, y= r \sin \theta$より、

（右辺）$= \cos \theta \times ( 2x ) + \sin \theta \times ( 2y ) = = \cos \theta \times ( 2r \cos \theta) + \sin \theta \times ( 2r sin \theta ) = 2r \times ( \cos^2 \theta + \sin^2 \theta ) = 2r$ 

### $\frac{\partial f}{\partial \theta} = -r \sin \theta \frac{\partial f}{\partial x} + r \cos \theta \frac{\partial f}{\partial y}$について

$f$は$\theta$に依存していないから、

（左辺）$= 0$

$x = r \cos \theta, y= r \sin \theta$より、

（右辺）$= -y \frac{\partial f}{\partial x} + x \frac{\partial f}{\partial y} = -y \times (2x) + x \times ( 2y ) = 0$
